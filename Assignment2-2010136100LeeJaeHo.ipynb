* 키보드로 정수값을 입력 받고 그 값이 양수인지 음수인지를 비트 연산자를 이용하여 판단하는 프로그램을 작성하시오.
    * [참고] 가장 왼쪽에 있는 비트가 0이면 양수이며, 1이면 음수이다.
    * [참고] 수업보충자료 (https://github.com/bluebibi/python-e-learning/blob/master/supplement.ipynb) 를 참고하시오.

```python
def findNegative(n):
    # 0x8 값을 이동시킬 변수 생성
    shift = n.bit_length()/4 + 1
    shift = 0x8 << shift*4
    
    # 0x8000 값을 통해 입력받은 정수의 첫 비트가 1 인지 판단한다.
    if ( ( n & shift ) == shift ):
        print "입력하신 값 %d은 음수입니다." % ( n )
    else:
        print "입력하신 값 %d은 양수입니다." % ( n )
        
num = input('정수를 입력하세요.:')
findNegative(num)
```

* 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
    * 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
    * 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
    * 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
    * 반환하는 리스트는 정렬이 되어 있어야 한다.
    * 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
        * list_union([1, 2, 3], [1, 2, 4])
        * [1, 2, 3, 4]
        * list_union([-10, -5, 0, -1], [100, 9, 0, 9])
        * [-10, -5, -1, 0, 9, 100]
        * list_union([0, 1, 2], [0, 1, 2])
        * [0, 1, 2]
    * [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
    * [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista, listb):
    # b리스트의 원소들을 하나씩 비교
    for y in listb :
        # a와 중복되지 않으면 추가
        if not ( y in lista ):
            lista.append(y)
    return sorted(lista)
print list_union([1, 2, 3], [1, 2, 4])
print list_union([-10, -5, 0, -1], [100, 9, 0, 9])
print list_union([0, 1, 2], [0, 1, 2])
```

* 앞에서부터 읽을 때나 뒤에서부터 읽을 때나 모양이 같은 수를 대칭수(palindrome)라고 부릅니다.
* 두 자리 수를 곱해 만들 수 있는 대칭수 중 가장 큰 수는 9009 (= 91 × 99) 입니다.
* 세 자리 수를 곱해 만들 수 있는 가장 큰 대칭수는 얼마입니까?

```python
def cubePalindrome():
    return sorted ( [ x * y for x in range(100,1000) for y in range(100, 1000) if ( str( x * y ) == str ( x * y )[::-1] ) ] )
l = cubePalindrome()
print l
print " 가장 큰 대칭수는 %d 입니다. " % (l[-1])
```

###### cubePalindrome이라는 함수는 세자리 정수 2개를 곱해서 그것이 대칭수이면 리스트에 넣어 리스트를 최종적으로 정렬해 반환해주는 함수 이다. 나는 함수 내부에서 리스트 내포를 사용했다. 하지만 리스트 내포를 사용하니 곱셈이 3번이 나와야 하기 때문에 수행속도에서 조금 느리게 코딩이 되었을 거라고 생각이 든다. 그럼에도 불구하고 리스트 내포를 사용한 이유는 이번에 배운 리스트 내포에 대해서 이해를 해보고 싶어 리스트 내포를 사용해 함수를 작성하였다.

* 1 ~ 10 사이의 어떤 수로도 나누어 떨어지는 가장 작은 수는 2520입니다.
* 그러면 1 ~ 20 사이의 어떤 수로도 나누어 떨어지는 가장 작은 수는 얼마입니까?

```python
# 소수 판별
def PrimaryN( n ):
    if n is 2:
        return True
    
    elif n > 2:
        for y in xrange (2, n):
            if ( n%y is 0 ):
                break
        else:
            return True
    return False

#소인수를 넣는 부분
def GetPrimary( n ):
    # 값을 저장하는 리스트
    result = []

    for z in xrange(2, n) :
        if n % z == 0 :
            if PrimaryN( z ):
                result += [z]
                n /= z
                result += GetPrimary( n )
                break
    else :
        if PrimaryN( n ) :
            result += [n]
            
        
    return result 
# 최소공배수를 구하는 부분
def getLCM( a, b ):
    result = 1
    # a 값에 대한 소인수 
    pA = GetPrimary(a)
    # b 값에 대한 소인수
    pB = GetPrimary(b)
    
    # a 와 b 의 소인수의 합집합에 대해 루프를 돌린다.
    for x in list( set(pA) | set(pB) ):
        if (pA.count(x) > pB.count(x)):
            result *= x ** pA.count(x)
        else:
            result *= x ** pB.count(x)
    
    return result

# 정해진 수까지 모든 수로 나누어 떨어지는 가장 작은 수를 구하는 함수
def getAllNumDivide( n ):
    x = 1
    y = 2
    # 2 number의 최소공배수를 계속해서 구해 나간다.
    while ( not( y == n + 1 ) ):
        x = getLCM( x, y )
        y += 1
    return x

print "1 ~ 10까지의 어떤 수로도 나누어 떨어지는 가장 작은 수는 %d" % (getAllNumDivide(10))
print "1 ~ 20까지의 어떤 수로도 나누어 떨어지는 가장 작은 수는 %d" % (getAllNumDivide(20))
```

###### 1 ~ 20까지의 어떤 수로도 나누어 떨어지는 가장작은 수라는 표현은 사실 1 ~ 20까지의 모든 수의 최소 공배수를 구하는 것과 같다. 이 문제를 풀기위해 어떤 값의 소인수를 얻는 함수를 만들고 두 정수를 인자로 받아 두 정수의 소인수를 구해 최소 공배수를 얻는 함수를 만들었다. 또한 그것을 통해 1 ~ 20까지 모든 수를 2개씩 비교해 가며 최소공배수를 얻어 마지막으로 얻은 최소공배수를 출력해 주었다.

* 1부터 10까지 자연수를 각각 제곱해 더하면 다음과 같습니다 (제곱의 합).
    * 12 + 22 + ... + 102 = 385
* 1부터 10을 먼저 더한 다음에 그 결과를 제곱하면 다음과 같습니다 (합의 제곱).
    * (1 + 2 + ... + 10)2 = 552 = 3025
* 따라서 1부터 10까지 자연수에 대해 "합의 제곱"과 "제곱의 합" 의 차이는 3025 - 385 = 2640 이 됩니다.
* 그러면 1부터 100까지 자연수에 대해 "합의 제곱"과 "제곱의 합"의 차이는 얼마입니까?

```python
#합의 제곱을 구하는 함수
def getSumPower (n):
    result = 0
    for x in xrange(n+1):
        result += x
    return pow(result, 2)
#제곱의 합을 구하는 함수
def getPowerSum (n):
    if n == 1:
        return 1
    else :
        return n**2 + getPowerSum(n-1)

print "1 ~ 10 자연수에 대해 합의 제곱은 %d, 제곱의 합은 %d이며, 둘의 차이는 %d이다." % (getSumPower(10), getPowerSum(10), getSumPower(10) - getPowerSum(10))
print "1 ~ 20 자연수에 대해 합의 제곱은 %d, 제곱의 합은 %d이며, 둘의 차이는 %d이다." % (getSumPower(20), getPowerSum(20), getSumPower(20) - getPowerSum(20))
```
