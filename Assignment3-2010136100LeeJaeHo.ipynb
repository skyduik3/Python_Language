```python
#primary값을 저장할 리스트 생성
primaryL = []

while 1 :
    #첫 번째 값
    primary = 2
    for x in primaryL:
        if primary % x == 0 : break
    else : 
        primaryL.append(primary)
    
    primary += 1
    if len(primaryL) == 10001: break

print "10001번째의 소수는 %d입니다." % (primaryL[-1])
```

* 예를 들면 32 + 42 = 9 + 16 = 25 = 52이므로 3, 4, 5는 피타고라스 수입니다.

* a + b + c = 1000 인 피타고라스 수 a, b, c는 한 가지 뿐입니다. 이 때, a × b × c 는 얼마입니까?

```python
import math

for x in range(1, 1000):
    for y in range(x, 1000):
        if( (x+y+math.sqrt(pow(x, 2) + pow(y, 2))) == 1000 ):
            print "a + b + c == 1000일때 a * b * c = %d" % ( x*y*math.sqrt(pow(x, 2) + pow(y, 2)) )
            break
    else: continue
    break
```
