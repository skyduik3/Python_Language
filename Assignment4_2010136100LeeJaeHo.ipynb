range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

* 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
        range(stop)
        range(start, stop[, step])
* 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
* 3) frange 사용 예
    * 3-1) frange(0.5)
        * [0.0, 0.1, 0.2, 0.3, 0.4]
    * 3-2) frange(1.0, 2.0)
        * [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    * 3-3) frange(2.2, 4.0, 0.5)
        * [2.2, 2.7, 3.2, 3.7]

```python
def frange (*arg):
    l = []
    
    if len(arg) == 1:
        start = 0.0
        stop = arg[0]
        step = 0.1
    elif len(arg) == 2:
        start = arg[0]
        stop = arg[1]
        step = 0.1
    elif len(arg) == 3:
        start = arg[0]
        stop = arg[1]
        step = arg[2]
        
    while(start < stop):
        l.append('{0:.1f}'.format(start))
        start += step
        
    return l

print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

값이 1개 2개 3개가 들어오는 경우에 따라서 start, stop, step 값을 조정해 준 후 frange()함수를 만들었다.

가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

* sum() 사용 예
    * sum()
        * 0
    * sum(1, 2)
        * 3
    * sum(1, 2, 3, 4, 5)
        * 15
    * sum(1, 5, 7, 2, -10)
        * 5

```python
def sum(*arg):
    result = 0 
    for i in arg:
        result += i
    return result

print sum()
print sum(1, 2)
print sum(1, 2, 3, 4, 5)
print sum(1, 5, 7, 2, -10)
```

[6번 문제]
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

* [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if n == 1:
        return n
    else :
        return n * myfact(n-1)
    
print myfact(1000)
```

양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

    n → n / 2 (n이 짝수일 때)
    n → 3 n + 1 (n이 홀수일 때)

13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.

    13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 
(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?

참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.

```python
mostNum = 0
for x in xrange(1, 1000001):
    count = 0
    while (x != 1):
        if x%2 == 0 :
            x = x/2
        else:
            x = 3*x + 1
        count += 1
    if count > mostNum:
        mostNum = count

print mostNum
```

1000000이하의 숫자의 루프를 돌리면서, 그 값들에 대해 우박수를 구해준다. 계산 과정을 거칠때마다 count를 카운트 해준다.

아래와 같은 2 × 2 격자의 왼쪽 위 모서리에서 출발하여 오른쪽 아래 모서리까지 도달하는 길은 모두 6가지가 있습니다 (거슬러 가지는 않기로 합니다).

```python
import math
def findRoad(n):
    width = n
    height = n
    total = width + height
    print math.factorial(total) / (math.factorial(width) * math.factorial(height))

findRoad(20)
```

```python
import urllib2
import string
import json
import pickle
import os

def ParsingPage(ss):
    source = urllib2.urlopen(ss).read()

    # Html Source Save
    f = open(ss.split('//')[1]+'.html', 'w')
    f.write(source)
    f.close()
    
    #print source # 추후 숙제 답안 제출 때엔 왼쪽 라인은 주석처리할 것
    
    # 먼저 '<'를 기준으로 나누어 준다.
    l = source.split('<')
    
    # 그다음 각각의 나뉘어진 원소에 대해서 '>'를 기준으로 나누어 2개 이상으로 나누어 졌다면, 오른쪽에 있는 문자열을 저장하고 그렇지 않으면 삭제한다.
    for x in range(len(l)):
        tempList = l[x].split('>')
        if len(tempList) > 1:
            # 구두 문자를 제거하고 값을 추가한다.
            tempString = tempList[-1]
            for punc in tempList[-1]:
                if ( punc in string.punctuation ):
                    tempString = tempString.replace(punc, '')
            l[x] = tempString
        else :
            l[x] = ' '
    
    # 모든 리스트의 값들을 다시 이어 붙인다.
    source = ' '.join(l)
    
    # dictionery 정의
    #불용어 제거
    negativeEnglish = ['a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']                           
    
    d = {}
    for x in source.split():
        if x.lower() not in negativeEnglish :
            if d.has_key(x):
                d[x] += 1
            else:
                d[x] = 1    
    
    # 출력한다.
    # 사전 객체 저장
    f = open(ss.split('//')[1]+'.words.frequency', 'w')
    pickle.dump(d, f)
    f.close()
    
    
ParsingPage('http://learningenglish.voanews.com')
ParsingPage('http://repeatafterus.com')
ParsingPage('http://www.audioenglish.org')
ParsingPage('http://www.esl-lab.com')
ParsingPage('http://www.npr.org')


```
